## Ejercicio 00

Muestre las principales características del dataset utilizando pandas.

En particular muestre los nombres de las columnas y sus tipos; el índice del dataset; la forma del dataset.

```python
def mostrar_caracteristicas_dataset(dataframe):
    """
    Muestra las principales características del dataset utilizando pandas.

    Parameters:
    dataframe (DataFrame): El dataframe que se va a analizar.

    Returns:
    None
    """
    # Muestra los nombres de las columnas y sus tipos
    print("Nombres de las columnas y sus tipos:")
    print(dataframe.dtypes)

    # Muestra el índice del dataset
    print("\nÍndice del dataset:")
    print(dataframe.index)

    # Muestra la forma del dataset
    print("\nForma del dataset:")
    print(dataframe.shape)

# Llamar a la función para mostrar las características del dataset
mostrar_caracteristicas_dataset(df)
```

## Ejercicio 01

Cree el df_01 con las siguientes características:

Diez columnas (dentro de las diez columnas deben estar la primera y la última columna del dataset original).

Diez filas (las filas deben corresponder a las cinco primeras filas y las cinco últimas filas del dataset original).

Muestre el resultado y las características del DataFrame.

```python
import pandas as pd

# Seleccionar las columnas de interés para df_01
columnas_interes_df01 = ["Driver"] + list(df.columns[-9:])

# Obtener las filas de interés para df_01
filas_interes = pd.concat([df.head(5), df.tail(5)])

# Crear df_01 con las diez columnas y diez filas
df_01 = filas_interes[columnas_interes_df01]

# Mostrar el resultado y las características del DataFrame df_01
print("DataFrame df_01:")
print(df_01)

# Mostrar las características del DataFrame df_01
print("\nCaracterísticas del DataFrame df_01:")
mostrar_caracteristicas_dataset(df_01)
```

## Ejercicio 06

Seleccione una columna del DataFrame **original** que contenga una variable categórica (si no hay variables categóricas cree una columna con variables categóricas a partir de alguna columna numérica del DataFrame).

Utilice el método **groupby** para analizar la información del DataFrame a partir de la variable categórica seleccionada.

Muestre el resultado y las características del DataFrame.

```python
# Verificar las columnas del DataFrame original para seleccionar una variable categórica
print("Columnas del DataFrame original:")
print(df.columns)

# Seleccionar una columna que contenga una variable categórica
columna_categorica = "Nationality"  # Seleccionar la columna categórica

# Utilizar el método groupby para analizar la información del DataFrame a partir de la variable categórica seleccionada
grupo_categorico = df.groupby(columna_categorica)

# Mostrar el resultado del análisis utilizando el método groupby
for grupo, datos in grupo_categorico:
    print(f"\nGrupo: {grupo}")
    print(datos)

# Mostrar las características del DataFrame original
print("\nCaracterísticas del DataFrame original:")
mostrar_caracteristicas_dataset(df)

```

## Ejercicio 07

Cree una tabla dinámica, utilizando el método **pivottable**, para analizar la información del DataFrame **original**.

Muestre el resultado y las características del DataFrame.

```python
# Utilizar el método pivot_table para crear una tabla dinámica y analizar la información del DataFrame original
tabla_dinamica = pd.pivot_table(df, index="Nationality", values="Championships", aggfunc="sum")

# Mostrar el resultado de la tabla dinámica
print("Tabla Dinámica:")
print(tabla_dinamica)

# Mostrar las características de la tabla dinámica
print("\nCaracterísticas de la Tabla Dinámica:")
print(tabla_dinamica.info())
```

## Ejercicio 08

Encuentre todos los valores nulos del DataFrame original y cambie esos valores nulos con la media de la respectiva columna.

Muestre el resultado y las características del DataFrame.

```python
# Encontrar todos los valores nulos del DataFrame original
valores_nulos = df.isnull().sum()

print("Valores nulos por columna:")
print(valores_nulos)

# Reemplazar los valores nulos con la media de la respectiva columna
df_reemplazado = df.fillna(df.mean(numeric_only=True))

# Mostrar el resultado del DataFrame con los valores nulos reemplazados
print("\nDataFrame con valores nulos reemplazados:")
print(df_reemplazado)

# Mostrar las características del DataFrame con valores nulos reemplazados
print("\nCaracterísticas del DataFrame con valores nulos reemplazados:")
mostrar_caracteristicas_dataset(df_reemplazado)


```
